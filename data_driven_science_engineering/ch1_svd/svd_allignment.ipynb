```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.misc
import os
plt.rcParams['figure.figsize'] = [8,8]
plt.rcParams.update({'font.size': 18})

n = 1000
q = int(n/4)
X = np.zeros((n,n))
X[(q-1):(3*q), (q-1):(3*q)] =1

plt.imshow(256-X)
plt.set_cmap('gray')
plt.axis('off')
plt.title('OG Matrix')
plt.show()
```
